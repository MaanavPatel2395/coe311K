# Make your own Least Squares Method

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
def My_lstsq(x, y):
    x_sum = np.sum(x)
    y_sum = np.sum(y)
    x_sumsq = np.sum(x**2)
    xy_sum = np.sum(x*y)
    n = len(x)
    m = (n*xy_sum - x_sum*y_sum) / (n*x_sumsq - (x_sum)**2)
    c = (y_sum - m*x_sum) / n
    return m, c
```

```python
x = np.array([random.uniform(0, 1) for i in range(100)])
y = np.array([random.uniform(0, 1) for i in range(100)])
x.sort()
```

```python
Y = np.zeros_like(y)
Y = [5.6*x[i] + 0.69 + random.uniform(-1, 1) for i in range(len(y))]
X = np.arange(0, 1, 0.01)
```

```python
A = np.vstack([X, np.ones(len(X))]).T
m, c = My_lstsq(X, Y)
M, C = np.linalg.lstsq(A, Y, rcond=None)[0]
print(m, c)
print(M, C)
y = m*x + c
```

My Function and the built in function have pretty close slope and intercept

```python
plt.plot(x, y)
plt.scatter(X, Y)
plt.title("Least Squares Fit")
plt.grid(True)
plt.show()
```

```python
errors = np.abs(Y - y)
print(errors, "\n")
print(errors.max(), errors.min())
```
