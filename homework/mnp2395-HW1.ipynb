# MonteCarlo Method Hw1

MonteCarlo function and imports:

```python
import numpy as np
import math
import random
import matplotlib.pyplot as plt

def approxPi(samples):
    counter = 0
    for i in range(0, samples):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        n = (x**2 + y**2)
        if (n<1):
            counter = counter + 1
    Circle = counter / samples
    APi = Circle*4
    return abs(math.pi - APi)
```

```python
approxPi(100000)
```

Making the x and y arrays for plotting

```python
x = np.arange(10, 1000, 20)
y = np.array([approxPi(n) for n in x]) #used AI to find out how to itterate an array through a function with an integer attribute
```

```python
plt.subplot(1, 1, 1)
plt.scatter(x, y)
plt.xlabel("Samples for MonteCarlo")
plt.ylabel("Error (Pi-ApproxPi)")
plt.title("Monte Carlo Error")
plt.show()
```

Conclusion:
There seems to be a downward trend as the number of samples (x) in the MonteCarlo function increases, meaning that the error between real Pi and the approximation decreases. This shows that the MonteCarlo function is more accurate at predicting Pi as you give it more samples (something I was expirementing with after writing the function *see above*)
