# Approximating sin(x)

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def sinTS(x, terms):
    res = 0
    for i in range (0, terms+1):
        res += (((-1)**i)*(x**(2*i+1)))/math.factorial(2*i+1)
    return res
```

```python
x = np.arange(0,10.1,0.1)
```

```python
y1 = sinTS(x, 1)
y2 = sinTS(x, 2)
y3 = sinTS(x, 3)
y4 = sinTS(x, 4)
```

```python
plt.scatter(x, y1, color='red')
plt.scatter(x, y2, color='orange')
plt.scatter(x, y3, color='yellow')
plt.scatter(x, y4, color='green')
plt.show()
plt.xlim(0,5*np.pi)
plt.ylim(-2, 2)
plt.scatter(x, y1, color='red')
plt.scatter(x, y2, color='orange')
plt.scatter(x, y3, color='yellow')
plt.scatter(x, y4, color='green')
plt.show()
```

Conclusion: The relative error from the sine function as the x value gets farther from 0 (as see by the function reaching into the hundreds when sine only goes to 1) both alternates between a positve and negative error (since polynomial switch as you change the sine of the largest term). It shows that the error grows exponentially as you move away from 0.
