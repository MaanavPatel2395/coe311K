# Quantifying Error

```python
import numpy as np
import math
import random
import matplotlib.pyplot as plt

def approxPi(samples):
    counter = 0
    for i in range(0, samples):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        n = (x**2 + y**2)
        if (n<1):
            counter = counter + 1
    Circle = counter / samples
    APi = Circle*4
    return abs(math.pi - APi)
```

```python
AbsError = ([approxPi(1000) for i in range(100)])
RelError = ([i/math.pi for i in AbsError])
```

```python
x = np.arange(0,100,1)
plt.plot(x, AbsError, 'b')
plt.plot(x, RelError, 'r')
plt.show()
```

```python
plt.scatter(x, AbsError)
plt.scatter(x, RelError)
plt.show()
```

Conclusion: The relative error of the MonteCarlo function is smaller than the absolute error because the relative error is a percentage difference from real Pi, while absolute error is a difference. If the numbers were bigger (3000 instead of 3.14, for example), the difference between absolute and relative errors will grow larger.
