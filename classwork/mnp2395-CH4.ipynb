# Leibniz Series for ùúã

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def leibniz(terms):
    res=0
    for n in range(0,terms+1):
        res += 4*(((-1)**n)/(2*n + 1))
        if (n%100==0):
            print(res, np.abs(np.pi - res))
    return res
```

```python
leibniz(5000)
```
