```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
v1 = np.array([1, 3, 5, 7])
v2 = np.array([2, 4, 6, 8])
```

```python
v_add = v1 + v2
v_sub = v1 - v2
print (v_add, v_sub)
```

```python
magnitude = np.linalg.norm(v_add)
print(magnitude)
```

```python
A = np.array([[1, 2, 3], [4, 5, 6]])
B = np.array([[2, 4, 6], [7, 8, 9], [1, 2, 5]])
print(A)
print(B)
```

```python
A = np.array([[2, 1], [1, 2]])
e_value, e_vector = np.linalg.eig(A)
print(e_value)
print(e_vector)
```

```python
A = np.array([[4, 1], [6, 3]])
A_inverse = np.linalg.inv(A)
identity = np.dot(A, A_inverse)
print(identity)
```
