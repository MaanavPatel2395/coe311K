```python
import numpy as np
import time
```

Define functions:

```python
def jacobi(A, b, x0, tol, max_iterations):
    start = time.perf_counter()
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
        # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            runtime = time.perf_counter() - start
            return x_new, k, runtime
        x = x_new
    runtime = time.perf_counter() - start
    return x, max_iterations, runtime
```

```python
def gauss_seidel(A, b, x0, tol, max_iterations):
    start = time.perf_counter()
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = x.copy()
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i)) # Using already updated values
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n)) # Using old values
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
            # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            runtime = time.perf_counter() - start
            return x_new, k, runtime
        x = x_new
    runtime = time.perf_counter() - start
    return x, max_iterations, runtime
```

Challenge 6a:

```python
A6 = np.array([[10.0, -1.0, 2.0, 0.0, 0.0],[-1.0, 11.0, -1.0, 3.0, 0.0],[2.0, -1.0, 10.0, -1.0, 0.0],[0.0, 3.0, -1.0, 8.0, -2.0], [0.0, 0.0, 0.0, -2.0, 9.0]])
B6 = np.array([14.0, 30.0, 26.0, 25.0, 37.0])
sol = np.zeros_like(B6)
```

```python
x, iteration, run = jacobi(A6, B6, sol, 0.000001, 100)
print(x, iteration, run)
x1, iteration1, run1 = gauss_seidel(A6, B6, sol, 0.000001, 100)
print(x1, iteration1, run1)
```

Challenge 6b:

```python
Ab6 = np.array([[1.0, 2.0, 3.0, 0.0, 0.0],[2.0, 1.0, 2.0, 3.0, 0.0],[3.0, 2.0, 1.0, 2.0, 3.0],[0.0, 3.0, 2.0, 1.0, 2.0], [0.0, 0.0, 3.0, 2.0, 1.0]])
Bb6 = np.array([14.0, 22.0, 33.0, 26.0, 22.0])
sol = np.zeros_like(Bb6)
```

```python
x3, iteration3, run3 = jacobi(Ab6, Bb6, sol, 0.000001, 100)
print(x3, iteration3, run3)
x4, iteration4, run4 = gauss_seidel(Ab6, Bb6, sol, 0.000001, 100)
print(x4, iteration4, run4)
```

Conclusion: If the matrix is not diagonally dominant, the jacobi and gauss_seidel methods do not work since you must have it so that the linear equation you are solving for has a big number you divide with while the other numbers do not have a big effect on the estimate you get. When the matrix is like 6b and is not dominant, the error does not converge and the estimated solutions grow bigger and bigger, leading to an infinite number of iterations.

Additionally, the runtime for the gauss method in both converging and diverging senarios is faster than jacobi
